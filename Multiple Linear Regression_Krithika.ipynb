### Creating dummies 

```python
bikeshare['season'].replace({1:'spring',2:'summer',3:'fall',4:'winter'},inplace=True)
bikeshare['yr'].replace({0:'2018',1:'2019'},inplace=True)
bikeshare['mnth'].replace({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'June',7:'July',8:'Aug',9:'Sept',10:'Oct',11:'Nov',12:'Dec'},inplace=True)
bikeshare['holiday'].replace({0:'No Holiday',1:'Holiday'},inplace=True)
bikeshare['weekday'].replace({0:'Mon',1:'Tue',2:'Wed',3:'Thurs',4:'Fri',5:'Sat',6:'Sun'},inplace=True)
bikeshare['workingday'].replace({0:'workingday',1:'Holiday_w'},inplace=True)
bikeshare['weathersit'].replace({1:'Clear',2:'Cloudy',3:'Light Rain',4:'Heavy Rain'},inplace=True)
```

```python
bikeshare.head()
```

```python
dummy=pd.get_dummies(bikeshare[['yr','holiday','workingday','season','mnth','weekday','weathersit']],drop_first=True,dtype= int)
dummy.head()
```

```python
full=pd.concat([bikeshare,dummy],axis=1)
full.head()
```

```python
full=full.drop(['season','yr','holiday','mnth','workingday','weekday','weathersit'],axis=1)
full.head()
```

```python
full.shape
```

```python
Checking whether the data in our model is violating any of the linear regression assumptions. The assumptions are:
1. Linear Relationship between x and y
The scatter plot between the numerical variables and also made a co-relation heat map to see 
if there is actually any relationship before we start building our model.
2. Error terms are normall distributed
A distplot for y_train  y_pred points. They were all normally distributed with their mean equal to 0.
3. Error terms have constant variance
A scatter plot of X_train vs residuals and all the residuals were ranging from -0.3 to 0.4. Thus, they have a constant variance.
4. Error terms are independent of each other
When we plotted the scatter plot of residuals, we did not find any pattern or clustering or any other continous pattern 
which concludes that they are not dependent on each other.
```
